RANDOM VARIABLES
    - map outcome of random expression to numerical values
TYPES
    DISCRETE -  take countable no of values
    CONTINUOUS -  take value from continuous range
PROBABILITY MASS FUNCTION(PMF)
    - probability distribution of discrete random variables
PROBABILITY DENSITY FUNCTION(PDF)
    - probability distribution of continuous random variables
Expectation E(X)
    -weighted average of random variable possible value
VARIANCE
    - measure the spread of random variables
STANDARD DERIVATION
    - Square root of variance

```python
#example
import numpy as np
outcome = np.array([1,2,3,4,5,6])   #random variable: dice roll
probability = np.array([1/6]*6)
expectation = np.sum(outcome*probability)   #expectation
print(expectation)
var = np.sum((outcome - expectation)**2 * probability)   #variance and standard derivation
std = np.sqrt(var)
print(var)
```

```python
# simulate dice roll and calculate probabilities
import numpy as np #simulating 10000 dice roll
rolls = np.random.randint(1,7,size = 10000)
#calculate probability
p_even = np.sum(rolls % 2 == 0)/len(rolls)
p_greater_than_4 = np.sum(rolls> 4)/len(rolls)
print(p_even)
print(p_greater_than_4)
```

```python
#create and analyze random variables
import matplotlib.pyplot as plt
from scipy.stats import uniform
#discrete random variable : dice rolls
outcomes = [1,2,3,4,5,6]
probability = [1/6]*6
plt.bar(outcomes, probability, color ="green", alpha =0.6)
plt.title("PMF of a dice roll")
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.show()
```

```python
x = np.linspace(0,1,100)
pdf = uniform.pdf(x,loc =0, scale =1)
plt.plot(x, pdf, color ="red")
plt.title("PDF of uniform(0,1)")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

COMMON PROBABILITY DISTRIBUTION
  -gaussian - bell shaped curved characterised by mean and standard derivation
  -binominal - model the no of successes in n independent bernoulli trial used in logistic regression
  -poisson - model the no of event in a fixed interval
  -uniform - equal probability for all outcome in range
VISUALIZING DISTRIBUTION AND UNDERSTANDING THEIR PROPERTIES
      SKEWNESS 
         - measure of symmetry
         - positive - tail on right
         - negative - tail on left
      KURTOSIS
         - measure of tailedness of distribution
         - high - more outlier
         - low - few outlier

```python
#examples
import numpy as np
import seaborn as sns
from scipy.stats import norm , binom, poisson, uniform
#gaussian or normal
x = np.linspace(10,20,1000)
plt.plot(x, norm.pdf(x, loc =0,scale =1), label ="Gaussian")

#binominal
n , p = 14,0.5
x = np.arange(0, n+1)
plt.bar(x, binom.pmf(x,n,p), alpha = 0.7, label="binomial")

#poisson
lam = 3
x = np.arange(0,10)
plt.bar(x, poisson.pmf(x,lam), alpha =0.8, label="poisson")
plt.title("Probability Distribution")
plt.legend()
plt.show()
```

```python
#uniform
x = np.random.uniform(low =0, high=10, size=1000)
sns.histplot(x, kde= True, label = "uniform", color="pink")
plt.show()
```

```python
#Analyxe dataset distribution
import matplotlib.pyplot as plt
import pandas as pd
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)
feature = df["sepal_length"]
print("Skewness:", feature.skew())
print("Kurtosis:", feature.kurtosis())
sns.histplot(feature, kde=True)
plt.title("Distribution of sepal length")
plt.show()
```
