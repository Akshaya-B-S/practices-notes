basic dataframe operation

```python
#viewing data
df.head()   #viewing first 5 data in dataframe by default , we can also specify how many we want to print
df.tail(3)  #this will display the last 3 data in dataframe
df.info()  #information summary of the dataframe
df.describe() #detailed statistical summary
```

#Selecting columns
df['Name']  #if u want to particular column
df[["Name","Age"]] if u want to two columns to display

#Filtering rows
print(df[df["Age]>25]) 

#selecting by position
print(df.iloc[0])  #if u want data item from first row by position
print(df.iloc[:,0])   #first column by postion

Selecting by label
print(df.loc[0])   first row by index label
print(df.loc[:,"Name"])   give name column data intem in this particular case
